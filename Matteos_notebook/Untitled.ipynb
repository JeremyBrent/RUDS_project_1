```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
#Import CSV to Dataframe
gdp_df = pd.read_csv("clean_gdp.csv")
gdp_df
```

```python
#Delete unwanted columns
del gdp_df['Unnamed: 0']
del gdp_df['Code']
gdp_df
```

```python
#Set new Columns
gdp_df.columns=['Country','Year','Hours','GDP','Population']
gdp_df
```

```python
#Make list of the countries to lookup and bring it up in DF
country_list = ["United States", "Mexico", "Venezuela", "Brazil", "South Africa", "Nigeria", "Poland", "Sweden", "Russia", "China", "South Korea", "Australia"]
gdp_df = gdp_df.loc[gdp_df['Country'].isin(country_list)]
gdp_df
```

```python
# Get Values for Years, hours and gdp for Multi y axiz plot
years = gdp_df['Year']
hours = gdp_df['Hours']
gdp = gdp_df['GDP']
```

```python
x = np.arange(1950, 2021, 1)
```

```python
#Create 2 y axis to plot GDP and Hours over Time

x = years
y1 = hours
y2 = gdp

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(x, y1, 'r-')
ax2.plot(x, y2, 'b-')

ax1.set_xlabel('Years')
ax1.set_ylabel('Hours', color='r')
ax2.set_ylabel('GDP', color='b')

plt.show()
```

```python
plt.figure(figsize=(20, 3))
plt.plot(years, gdp, color='green')
```

```python
plt.figure(figsize=(20, 5))
plt.bar(years, hours, color='r', alpha=0.2, align="center")
```
